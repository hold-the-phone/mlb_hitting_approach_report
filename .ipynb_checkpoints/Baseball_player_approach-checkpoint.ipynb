```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as PLT
from matplotlib import cm as CM
from matplotlib import mlab as ML

pd.options.display.max_columns=150
```

```python
df = pd.read_csv('savant_data.csv')
df['game_date'] = pd.to_datetime(df['game_date'])
df.head()
```

```python
'''
Types of pitch:
AB Automatic Ball
AS Automatic Strike
CH Change-up
CU Curveball
EP Eephus
FC Cutter
FF Four-Seam Fastball
FO Forkball
FS Splitter
FT Two-Seam Fastball (synonymous with SI)
GY Gyroball
IN Intentional Ball
KC Knuckle Curve
KN Knuckleball
NP No Pitch
PO Pitchout
SC Screwball
SI Sinker (synonymous with FT)
SL Slider
UN Unknown 
'''
```

```python
np.sort(df['player_name'].unique())
```

```python
PLAYER_NAME="Matt Chapman"
```

```python
player_df = df[df['player_name']==PLAYER_NAME]

player_df['pitch_type_fast'] = np.where(((player_df['pitch_type']=='FC')|
                                        (player_df['pitch_type']=='FF')|
                                        (player_df['pitch_type']=='FT')|
                                        (player_df['pitch_type']=='SI')),1,0)

player_df['pitch_type_offspeed'] = np.where(((player_df['pitch_type']=='CH')|
                                        (player_df['pitch_type']=='CU')|
                                        (player_df['pitch_type']=='EP')|
                                        (player_df['pitch_type']=='FO')|
                                        (player_df['pitch_type']=='FS')|
                                        (player_df['pitch_type']=='SC')|
                                        (player_df['pitch_type']=='SL')),1,0)

player_df['swing'] = np.where(((player_df['description']!='ball')&
                              (player_df['description']!='blocked_ball')&
                              (player_df['description']!='called_strike')),1,0)

player_df['hit'] = np.where(((player_df['events']=='single')|
                            (player_df['events']=='double')|
                            (player_df['events']=='triple')|
                            (player_df['events']=='home_run')),1,0)

player_df['out'] = np.where(((player_df['events']=='double_play')|
                            (player_df['events']=='field_out')|
                            (player_df['events']=='fielders_choice')|
                            (player_df['events']=='fielders_choice_out')|
                            (player_df['events']=='force_out')|
                            (player_df['events']=='grounded_into_double_play')|
                            (player_df['events']=='sac_fly')| #INCLUDE?
                            (player_df['events']=='strikeout')|
                            (player_df['events']=='strikeout_double_play')),1,0)

player_df
```

```python
np.mean(player_df['swing'])
```

#### build swing rate by location heatmap

```python
df_relevant = player_df[(player_df['pitch_type_fast']==1)&
                        (player_df['p_throws']=='R')]

x = df_relevant['plate_x']
y = df_relevant['plate_z']
z = df_relevant['swing']
X, Y = np.meshgrid(x, y)

gridsize=30
PLT.figure(figsize=(15,10))
PLT.subplot(111)

PLT.hexbin(x, y, C=z, gridsize=gridsize, cmap=CM.jet, bins=None)
PLT.axis([x.min(), x.max(), y.min(), y.max()])

cb = PLT.colorbar()
cb.set_label('mean value')

PLT.plot([0.958, 0.958], [0, 5], color='green', linestyle='-', linewidth=3)
PLT.plot([-0.958, -0.958], [0, 5], color='green', linestyle='-', linewidth=3)

PLT.plot([-2, 2], [1.60, 1.60], color='green', linestyle='-', linewidth=3)
PLT.plot([-2, 2], [3.55, 3.55], color='green', linestyle='-', linewidth=3)

PLT.xticks(np.arange(-3, 4, step=1))
PLT.yticks(np.arange(0, 5, step=1))

PLT.show()
```

#### Function for contact hit rate/out rate

```python
def contact_heatmap(result='hit',p_throws='R',fast_offspeed = 'fast'):
    '''
    result can be 'hit or 'out',
    p_throws can be 'R' or 'L',
    fast_offspeed can be 'fast' or 'offspeed'
    '''
    fastball_indicator=0
    if fast_offspeed=='fast':
        df_relevant = player_df[(player_df['pitch_type_fast']==1)&
                        (player_df['p_throws']==p_throws)&
                        (player_df['swing']==1)&
                        (player_df['events']!='strikeout')]
    else:
        df_relevant = player_df[(player_df['pitch_type_offspeed']==1)&
                        (player_df['p_throws']==p_throws)&
                        (player_df['swing']==1)&
                        (player_df['events']!='strikeout')]

    x = df_relevant['plate_x']
    y = df_relevant['plate_z']
    z = df_relevant[result]
    X, Y = np.meshgrid(x, y)
    
    gridsize=16
    PLT.figure(figsize=(15,10))
    PLT.subplot(111)
    
    PLT.hexbin(x, y, C=z, gridsize=gridsize, cmap=CM.jet, bins=None)
    PLT.axis([x.min(), x.max(), y.min(), y.max()])
    
    cb = PLT.colorbar()
    cb.set_label('mean value')
    
    PLT.plot([0.958, 0.958], [1, 4], color='green', linestyle='-', linewidth=3) #in/off plate
    PLT.plot([-0.958, -0.958], [1, 4], color='green', linestyle='-', linewidth=3)
    
    PLT.plot([-2, 2], [1.60, 1.60], color='green', linestyle='-', linewidth=3) #upper/lower strikezones
    PLT.plot([-2, 2], [3.55, 3.55], color='green', linestyle='-', linewidth=3)
    
    PLT.plot([0, 0], [1, 4], color='red', linestyle='--', linewidth=3) #red midline
    PLT.plot([-2, 2], [2.575, 2.575], color='red', linestyle='--', linewidth=3) #red midline
    
    PLT.xticks(np.arange(-3, 4, step=1))
    PLT.yticks(np.arange(0, 6, step=1))
    
    PLT.show()
    
```

#### Take a look at exit velo

```python
def exit_velo_heatmap(p_throws='R',fast_offspeed = 'fast'):
    
    '''
    result can be 'hit or 'out',
    p_throws can be 'R' or 'L',
    fast_offspeed can be 'fast' or 'offspeed'
    '''
    if fast_offspeed=='fast':
        df_relevant = player_df[(player_df['pitch_type_fast']==1)& #as opposed to pitch_type_offspeed
                        (player_df['p_throws']==p_throws)&
                        (player_df['swing']==1)&
                        (player_df['launch_speed']!='null')]
    else:
        df_relevant = player_df[(player_df['pitch_type_offspeed']==1)& #as opposed to pitch_type_offspeed
                        (player_df['p_throws']==p_throws)&
                        (player_df['swing']==1)&
                        (player_df['launch_speed']!='null')]

    
    x = df_relevant['plate_x']
    y = df_relevant['plate_z']
    z = df_relevant['launch_speed']
    X, Y = np.meshgrid(x, y)
    
    gridsize=12
    PLT.figure(figsize=(15,10))
    PLT.subplot(111)
    
    PLT.hexbin(x, y, C=z, gridsize=gridsize, cmap=CM.jet, bins=None)
    PLT.axis([x.min(), x.max(), y.min(), y.max()])
    
    cb = PLT.colorbar()
    cb.set_label('mean value')
    
    PLT.plot([0.958, 0.958], [1, 4], color='green', linestyle='-', linewidth=3) #in/off plate
    PLT.plot([-0.958, -0.958], [1, 4], color='green', linestyle='-', linewidth=3)
    
    PLT.plot([-2, 2], [1.60, 1.60], color='green', linestyle='-', linewidth=3) #upper/lower strikezones
    PLT.plot([-2, 2], [3.55, 3.55], color='green', linestyle='-', linewidth=3)
    
    PLT.plot([0, 0], [1, 4], color='red', linestyle='--', linewidth=3) #red midline
    PLT.plot([-2, 2], [2.575, 2.575], color='red', linestyle='--', linewidth=3) #red midline
    
    PLT.xticks(np.arange(-3, 4, step=1))
    PLT.yticks(np.arange(0, 6, step=1))
    
    PLT.show()
```

```python
contact_heatmap('hit',p_throws='R',fast_offspeed='fast')
contact_heatmap('out',p_throws='R',fast_offspeed='fast')

exit_velo_heatmap(p_throws='R',fast_offspeed = 'fast')
exit_velo_heatmap(p_throws='R',fast_offspeed = 'offspeed')
```
